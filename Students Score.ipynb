# Gender Distribution

```python
ax = sns.countplot(x = "Gender", data = df, width = 0.6)
plt.figure(figsize = (3,1))
plt.show
ax.bar_label(ax.containers[0])
```

from the above chart we have analysed that:
    no of females in data is more than the males

# Impact of parents education on students score

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean', "ReadingScore":'mean', "WritingScore":'mean'})
print(gb)
```

```python
sns.heatmap(gb, annot = True)
plt.show()
plt.figure(figsize = (3,4))
```

from the above chart we have concluded that:
    the education of parents have good impact on their students score

# Impact of ParentMaritalStatus on students score

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean', "ReadingScore":'mean', "WritingScore":'mean'})
print(gb1)
```

```python
sns.heatmap(gb1, annot = True)
plt.show()
plt.figure(figsize = (6,4))
```

from the above chart we have concluded that: the marital status of parents have not much impact on their students score
